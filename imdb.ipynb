```python
# Step 2: Write your Streamlit app to a file
%%writefile app.py
import streamlit as st
import requests

st.title("My GitHub Contributions")

# GitHub username
username = st.text_input("Enter your GitHub username:", "")

if username:
    # Fetch the user's repositories
    repos_url = f"https://api.github.com/users/{username}/repos"
    repos_response = requests.get(repos_url)
    repos = repos_response.json()

    if repos_response.status_code == 200:
        for repo in repos:
            st.subheader(repo['name'])
            st.write(repo['description'])
            st.write(f"[GitHub Repository]({repo['html_url']})")

            # Fetch the pull requests for the repository
            pulls_url = f"https://api.github.com/repos/{username}/{repo['name']}/pulls?state=all"
            pulls_response = requests.get(pulls_url)
            pulls = pulls_response.json()

            if pulls_response.status_code == 200:
                for pull in pulls:
                    st.write(f"- [{pull['title']}]({pull['html_url']}) (State: {pull['state']})")
            else:
                st.write("Could not fetch pull requests.")
    else:
        st.write("Could not fetch repositories.")

# Display the updates
st.subheader("Updates")
if 'updates' not in st.session_state:
    st.session_state['updates'] = []

with st.form("add_update"):
    new_update = st.text_area("Add a new update:")
    submitted = st.form_submit_button("Add Update")

    if submitted and new_update:
        st.session_state['updates'].append(new_update)

for update in st.session_state['updates']:
    st.write(f"- {update}")

# Step 3: Create a function to run Streamlit with ngrok
from pyngrok import ngrok

def run_streamlit():
    # Terminate any existing ngrok connections
    ngrok.kill()

    # Create a new ngrok tunnel
    public_url = ngrok.connect(port='8501')
    print(f"Streamlit app running at: {public_url}")

    # Run the Streamlit app
    !streamlit run app.py

run_streamlit()
```

```python
# Load model directly
from transformers import AutoTokenizer, AutoModelForSequenceClassification

tokenizer = AutoTokenizer.from_pretrained("Hate-speech-CNERG/hindi-abusive-MuRIL")
model = AutoModelForSequenceClassification.from_pretrained("Hate-speech-CNERG/hindi-abusive-MuRIL")
```

```python
import torch  # Import the torch module

inputs = tokenizer("tu kidhar hai bhai, chutiya jaldi aana fir", return_tensors="pt")

with torch.no_grad():
    logits = model(**inputs).logits
```

```python
predicted_class_id = logits.argmax().item()

# Define the labels for your model
labels = ["Non-Abusive", "Abusive"]  # Replace with your actual label names

# Get the actual label name
predicted_label = labels[predicted_class_id]
print(predicted_label)
```
